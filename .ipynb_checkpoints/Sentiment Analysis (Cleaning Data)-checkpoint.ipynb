## Function to clean the text

```python
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import ToktokTokenizer
from pattern.en import suggest
import re

import nltk
nltk.download('stopwords')
```

```python
#Reduce repitition of characters, because it is well known that 
#English words ussually has only 2 repeated characters
def ReduceLengthening(text):
  pattern = re.compile(r"(.)\1{2,}")
  return pattern.sub(r"\1\1", text)
```

```python
toktok = ToktokTokenizer().tokenize
stemmer = SnowballStemmer('english')

def CleanText(text):
  mentionPat = r'@\w+'
  linkPat = r'(http|https)?://[\w./]+'
  combinedPat = r'|'.join((mentionPat,linkPat))
  stop_words = set(stopwords.words('english'))
  
  #Clean text from mention and links
  withoutML = re.sub(combinedPat, '', text)
  
  #Clean text from unknown character (unciode = u\ufffd)
  withoutUC = withoutML.replace(u"\ufffd", "?")
  
  #Clean text from non text characters
  textCharOnly = re.sub("[^A-Za-z]"," ",withoutUC)
  
  #lowercase only
  lowercase = textCharOnly.lower()
  
  #reduce lengthening of word (amazzziiingg -> amazziingg)
  reducedLength = ReduceLengthening(lowercase)
  
  #spell checking         
  words = reducedLength.split()
  correctedWords = []
  for word in words:
    correctedWord = suggest(word)[0]
    if correctedWord[1]>=0.7:
        #if the correction suggestion is 70% close
        #to the current word then it deserved to be the correction
        correctedWords.append(correctedWord[0])
    else :
        correctedWords.append(word)

  #Tokenize
  tokenizedWords = toktok(reducedLength)

  #remove stopwords  
  withoutSW = [word for word in tokenizedWords if not (word in stop_words)]
  
  #stemming
  stemmed = [stemmer.stem(word) for word in withoutSW]
  
  clean = " ".join(stemmed)
  
  return clean
```

### Parallelize the data cleaning to make it fast enough

```python
from multiprocessing import cpu_count, Pool
cores = cpu_count()
partitions = cores
def Parallelize(data, func):
    data_split = np.array_split(data, partitions)
    pool = Pool(cores)
    data = pd.concat(pool.map(func, data_split))
    pool.close()
    pool.join()
    return data

def CleanData(data):
    data["Text"] = data["Text"].apply(CleanText)
    return data
```

### Apply the cleaning method to the tweets of dataTrain

```python
%%time
dataTrain= Parallelize(dataTrain, CleanData)
```

##Save the clean tweet to drive

```python
dataTrain.to_csv("clean_tweet_training.csv")
```

### Apply the cleaning method to the tweets of dataTest

```python
%%time
dataTest = Parallelize(dataTest, CleanData)
```
